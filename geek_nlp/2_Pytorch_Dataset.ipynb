## Overview

In this tutorial,we will cover the basics of dataset and iterators so taht we can train models using gradient descents.

The best tutorial can be found in the official websites(https://pytorch.org/tutorials/beginner/data_loading_tutorial.html) 

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class MyStupidDataset(Dataset):
    def __init__(self):
        self.data = torch.randn([1024, 10, 10])
    
    def __len__(self):
        return 1024
    
    def __getitem__(self, idx):
        return self.data[idx, :, :]
```

```python
my_stupid_dataset = MyStupidDataset()
```

```python
my_data_loader = DataLoader(my_stupid_dataset, batch_size=64, shuffle=True)
```

```python
for i in my_data_loader:
    print(i)
```

```python
class MyDictDataset(Dataset):
    def __init__(self):
        self.x = torch.randn(1024, 10)
        self.y = torch.randn(1024)
    
    def __len__(self):
        return 1024
    
    def __getitem__(self, idx):
        return {'x': self.x[idx, :], 'y':self.y[idx]} 
```
